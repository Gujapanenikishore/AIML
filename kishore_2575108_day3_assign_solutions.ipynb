# 1.write  a python program that copies a file reading and writing upto 50 characters at a time




```python
# Function to copy a file reading and writing up to 50 characters at a time
def copy_file(source_file, destination_file):
    try:
        with open(source_file, 'r') as source, open(destination_file, 'w') as destination:
            while True:
                # Read up to 50 characters from the source file
                data = source.read(50)
                
                # If there's no more data to read, break the loop
                if not data:
                    break
                
                # Write the read data to the destination file
                destination.write(data)
    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Usage example
source_file = "source.txt"  # Replace with the name of your source file
destination_file = "destination.txt"  # Replace with the name of your destination file

copy_file(source_file, destination_file)
```

# 2. print all the numbers in the text file and print the no of blank spaces

```python
# Function to print numbers from a text file and count blank spaces
def print_numbers_and_count_blanks(sample.txt):
    try:
        with open(sample.txt, 'r') as file:
            content = file.read()
            numbers = []
            blank_space_count = 0

            for word in content.split():
                if word.isdigit():
                    numbers.append(int(word))
                elif word.isspace():
                    blank_space_count += 1

            if numbers:
                print("Numbers in the file:", numbers)
            else:
                print("No numbers found in the file.")
            
            print("Number of blank spaces:", blank_space_count)

    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Usage example
file_name = "sample.txt"  # Replace with the name of your text file
print_numbers_and_count_blanks(file_name)
```

# 3.write a function called sed that takes as arguments a patternstring,a replacement string,and two filenames; it should read the first file and write  the contents into the  second file(creating it if necessasry).if the pattern string appears anywhere in the file ,if should be replaced with the reeplacement string.if an erroroccurs while opening,reading,writting, or closing files,your program should catch the exception,print an error message,and exit 

```python
def sed(pattern, replacement, source_file, target_file):
    try:
        # Open the source file for reading
        with open(source_file, 'r') as source:
            # Read the content of the source file
            content = source.read()

        # Perform the replacement
        modified_content = content.replace(pattern, replacement)

        # Open the target file for writing (creating it if necessary)
        with open(target_file, 'w') as target:
            # Write the modified content to the target file
            target.write(modified_content)

        print(f"Pattern '{pattern}' replaced with '{replacement}' in {source_file} and saved to {target_file}")

    except FileNotFoundError:
        print(f"Error: {source_file} not found.")
    except PermissionError:
        print(f"Error: Permission denied while accessing files.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
pattern = "old_string"
replacement = "new_string"
source_file = "input.txt"
target_file = "output.txt"

sed(pattern, replacement, source_file, target_file)
```

# 6.write a python script that concatenates the contents of multiple text files into a single output file .allow the user to specify the input files and the output file

```python
# Function to concatenate text files into a single output file
def concatenate_files(input_files, output_file):
    try:
        with open(output_file, 'w') as output:
            for input_file in input_files:
                with open(input_file, 'r') as file:
                    output.write(file.read())
        print(f"Files have been concatenated into {output_file}")
    except FileNotFoundError as e:
        print(f"File not found: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Get input file names from the user
num_files = int(input("Enter the number of input files: "))
input_files = []
for i in range(num_files):
    file_name = input(f"Enter the name of input file {i + 1}: ")
    input_files.append(file_name)

# Get the output file name from the user
output_file = input("Enter the name of the output file: ")

# Concatenate the input files into the output file
concatenate_files(input_files, output_file)
```

# 8.Assume that you are developing a student gradebook system for a school.The system should allow teachers to input student grades for various subjects,store the data in files, and provide students with the ability to view their grades .Design a python program thats accomplishes the following tasks

```python
#i)teachers should be able to input grades for students in different subjects
import os

# Define a dictionary to store student grades
gradebook = {}

# Function to input and store student grades
def input_grades():
    student_name = input("Enter student name: ")
    subject = input("Enter subject name: ")
    grade = input("Enter the grade: ")

    if student_name not in gradebook:
        gradebook[student_name] = {}
    if subject not in gradebook[student_name]:
        gradebook[student_name][subject] = []

    gradebook[student_name][subject].append(grade)
    print("Grade added successfully!")

# Function to save gradebook data to a file
def save_gradebook_to_file(filename):
    with open(filename, 'w') as file:
        for student, grades in gradebook.items():
            file.write(f"{student}:\n")
            for subject, subject_grades in grades.items():
                file.write(f"  {subject}: {', '.join(subject_grades)}\n")

# Function to display student grades
def display_grades(student_name):
    if student_name in gradebook:
        print(f"Grades for {student_name}:")
        for subject, subject_grades in gradebook[student_name].items():
            print(f"  {subject}: {', '.join(subject_grades)}")
    else:
        print(f"{student_name} not found in the gradebook.")

# Main program loop
while True:
    print("\nStudent Gradebook System")
    print("1. Input Grades")
    print("2. Save Gradebook")
    print("3. View Grades")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        input_grades()
    elif choice == '2':
        filename = input("Enter the filename to save the gradebook data: ")
        save_gradebook_to_file(filename)
    elif choice == '3':
        student_name = input("Enter student name to view grades: ")
        display_grades(student_name)
    elif choice == '4':
        break
    else:
        print("Invalid choice. Please try again.")

print("Exiting the program.")
```
