# 2.Virtual Environmental Management 
Scenario: You are working on multiple python projects with different dependencies  and versions.To avoid conflicts and ensure projects  iolation,you decide to use virtual environments.
Create  a python  program  that demonstrates the following:
1.create  a virtual environment for a specific project.
2.Activate  and deactivate  virtual  environments.
3.Install,upgrade, and uninstall packages within  a virtual environment.
4.List all installed packages in a virtual environmental
5.Implements error  handling for virtuals  environment operations

```python
import os
import sys
import subprocess

# Function to create a virtual environment for a specific project
def create_virtual_environment(venv_name):
    try:
        subprocess.check_call([sys.executable, '-m', 'venv', venv_name])
        print(f"Virtual environment '{venv_name}' created successfully.")
    except subprocess.CalledProcessError:
        print(f"Error: Failed to create virtual environment '{venv_name}'.")

# Function to activate a virtual environment
def activate_virtual_environment(venv_name):
    venv_path = os.path.join(venv_name, 'bin' if sys.platform != 'win32' else 'Scripts', 'activate')
    try:
        subprocess.check_call([venv_path])
        print(f"Activated virtual environment '{venv_name}'.")
    except subprocess.CalledProcessError:
        print(f"Error: Failed to activate virtual environment '{venv_name}'.")

# Function to deactivate the currently active virtual environment
def deactivate_virtual_environment():
    try:
        subprocess.check_call(['deactivate'], shell=True)
        print("Deactivated the current virtual environment.")
    except subprocess.CalledProcessError:
        print("Error: Failed to deactivate the virtual environment.")

# Function to install a package within a virtual environment
def install_package(venv_name, package_name):
    try:
        subprocess.check_call([os.path.join(venv_name, 'bin' if sys.platform != 'win32' else 'Scripts', 'pip'), 'install', package_name])
        print(f"Package '{package_name}' installed successfully in '{venv_name}'.")
    except subprocess.CalledProcessError:
        print(f"Error: Failed to install package '{package_name}' in '{venv_name}'.")

# Function to list all installed packages in a virtual environment
def list_installed_packages(venv_name):
    try:
        installed_packages = subprocess.check_output([os.path.join(venv_name, 'bin' if sys.platform != 'win32' else 'Scripts', 'pip'), 'list'])
        print(f"Installed packages in '{venv_name}':\n{installed_packages.decode()}")
    except subprocess.CalledProcessError:
        print(f"Error: Failed to list installed packages in '{venv_name}'.")

# Usage example
if __name__ == "__main__":
    venv_name = "my_project_env"

    # 1. Create a virtual environment
    create_virtual_environment(venv_name)

    # 2. Activate the virtual environment
    activate_virtual_environment(venv_name)

    # 3. Install, upgrade, and uninstall packages
    install_package(venv_name, "requests")
    install_package(venv_name, "numpy")
    install_package(venv_name, "matplotlib")
    list_installed_packages(venv_name)

    # 4. Deactivate the virtual environment
    deactivate_virtual_environment()
```

```python
CREATE DATABASE inventory;

USE inventory;

CREATE TABLE purchases (
    id INT AUTO_INCREMENT PRIMARY KEY,
    product_name VARCHAR(255),
    quantity INT,
    purchase_date DATE
);

CREATE TABLE sales (
    id INT AUTO_INCREMENT PRIMARY KEY,
    product_name VARCHAR(255),
    quantity INT,
    sale_date DATE
);

CREATE TABLE inventory (
    id INT AUTO_INCREMENT PRIMARY KEY,
    product_name VARCHAR(255),
    quantity INT
);
```

```python
import mysql.connector

# Connect to the MySQL database
connection = mysql.connector.connect(
    host="localhost",
    user="your_username",
    password="your_password",
    database="inventory"
)

cursor = connection.cursor()

# Function to insert data into the purchases table
def add_purchase(product_name, quantity, purchase_date):
    sql = "INSERT INTO purchases (product_name, quantity, purchase_date) VALUES (%s, %s, %s)"
    values = (product_name, quantity, purchase_date)
    cursor.execute(sql, values)
    connection.commit()
    print("Purchase added successfully.")

# Function to insert data into the sales table
def add_sale(product_name, quantity, sale_date):
    sql = "INSERT INTO sales (product_name, quantity, sale_date) VALUES (%s, %s, %s)"
    values = (product_name, quantity, sale_date)
    cursor.execute(sql, values)
    connection.commit()
    print("Sale added successfully.")

# Function to update inventory based on purchases and sales
def update_inventory():
    cursor.execute("DELETE FROM inventory")
    cursor.execute("""
        INSERT INTO inventory (product_name, quantity)
        SELECT product_name, SUM(quantity)
        FROM purchases
        GROUP BY product_name
    """)
    cursor.execute("""
        UPDATE inventory i
        INNER JOIN (
            SELECT product_name, SUM(quantity) as sold
            FROM sales
            GROUP BY product_name
        ) s ON i.product_name = s.product_name
        SET i.quantity = i.quantity - s.sold
    """)
    connection.commit()
    print("Inventory updated successfully.")

# Function to get current inventory
def get_inventory():
    cursor.execute("SELECT * FROM inventory")
    result = cursor.fetchall()
    return result

if __name__ == "__main__":
    # Example usage:
    add_purchase("Product A", 100, "2023-09-15")
    add_purchase("Product B", 50, "2023-09-16")
    add_sale("Product A", 20, "2023-09-17")
    update_inventory()
    inventory = get_inventory()
    print("Current Inventory:")
    for item in inventory:
        print(f"{item[1]}: {item[2]}")

    # Close the database connection
    cursor.close()
    connection.close()

```

# 2.Customer order processing 
Scenario: You are building  a customer order processing system for an e-commerce company.The system needs to interact with my sql database to store customer order,products and order details.
1.Design a MYSQL database schema for the order processing system.including tables for customers,products and orders.
2.write a python program that conects to the database and allows customers to place  new orders.
3.Implements  a feature that calculates that total cost of an orde and updates product quantities in the database
4.how would you handle cases where a product is no longer available when a customer places an order?
5.Develop a function to geenrate order reports for the company's finance department




```python
##1.Design a MYSQL database schema for the order processing system.including tables for customers,products and orders.
CREATE DATABASE order_processing;

USE order_processing;

CREATE TABLE customers (
    customer_id INT AUTO_INCREMENT PRIMARY KEY,
    customer_name VARCHAR(255),
    email VARCHAR(255)
);

CREATE TABLE products (
    product_id INT AUTO_INCREMENT PRIMARY KEY,
    product_name VARCHAR(255),
    price DECIMAL(10, 2),
    quantity INT
);

CREATE TABLE orders (
    order_id INT AUTO_INCREMENT PRIMARY KEY,
    customer_id INT,
    order_date DATE,
    total_cost DECIMAL(10, 2),
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);
```

```python
import mysql.connector

# Function to connect to the MySQL database
def connect_to_database():
    return mysql.connector.connect(
        host="localhost",
        user="your_username",
        password="your_password",
        database="order_processing"
    )

# Function to place a new order
def place_order(customer_id, product_id, quantity):
    try:
        connection = connect_to_database()
        cursor = connection.cursor()

        # Retrieve product information
        cursor.execute("SELECT product_name, price, quantity FROM products WHERE product_id = %s", (product_id,))
        product_info = cursor.fetchone()

        if not product_info:
            print("Error: Product not found.")
            return

        product_name, price, available_quantity = product_info

        if available_quantity < quantity:
            print(f"Error: Insufficient quantity of {product_name} available.")
            return

        total_cost = price * quantity

        # Update product quantity in the database
        cursor.execute("UPDATE products SET quantity = quantity - %s WHERE product_id = %s", (quantity, product_id))

        # Insert the order into the database
        cursor.execute("INSERT INTO orders (customer_id, order_date, total_cost) VALUES (%s, CURDATE(), %s)", (customer_id, total_cost))

        connection.commit()
        print(f"Order placed successfully. Total cost: ${total_cost:.2f}")

    except mysql.connector.Error as err:
        print(f"Database error: {err}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()

# Function to generate order reports
def generate_order_report():
    try:
        connection = connect_to_database()
        cursor = connection.cursor()

        cursor.execute("SELECT order_id, customer_name, order_date, total_cost FROM orders o JOIN customers c ON o.customer_id = c.customer_id")

        print("Order Report:")
        for order_id, customer_name, order_date, total_cost in cursor.fetchall():
            print(f"Order ID: {order_id}, Customer: {customer_name}, Order Date: {order_date}, Total Cost: ${total_cost:.2f}")

    except mysql.connector.Error as err:
        print(f"Database error: {err}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()

if __name__ == "__main__":
    # Example usage:
    place_order(1, 1, 3)  # Customer ID 1 places an order for 3 units of Product ID 1
    place_order(2, 2, 2)  # Customer ID 2 places an order for 2 units of Product ID 2
    generate_order_report()
```

```python
import mysql.connector

try:
    # 1. Define database connection parameters
    db_config = {
        "host": "localhost",
        "port": 3306,
        "user": "your_username",
        "password": "your_password",
        "database": "your_database"
    }

    # 2. Connect to the MySQL database
    connection = mysql.connector.connect(**db_config)

    # 3. Retrieve all records from the your_table table
    cursor = connection.cursor()
    cursor.execute("SELECT id, name, quantity FROM your_table")

    # Initialize total quantity
    total_quantity = 0

    # 4. Calculate the total quantity of all records retrieved
    for id, name, quantity in cursor.fetchall():
        total_quantity += quantity

    print(f"Total quantity before doubling: {total_quantity}")

    # 5. Update the quantity column by doubling its value
    cursor.execute("UPDATE your_table SET quantity = quantity * 2")

    # 6. Commit the changes to the database
    connection.commit()
    print("Quantity updated successfully.")

except mysql.connector.Error as err:
    print(f"Database error: {err}")

except Exception as e:
    print(f"Error: {e}")

finally:
    # 7. Close the database connection
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("Database connection closed.")
```
