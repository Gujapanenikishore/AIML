# RAILWAY MANAGEMENT SYSTEM


>Prorblem statement:
 The problem statement is to create the RAILWAY MANAGEMENT SYSTEM that develops a user- fiendly software  application that facilitates various functionalities related to railway ticketing and management .the system should allow users to book tickets,cancel bookings .check fares,view their bookings and display availiabale trains.

>case study:
Imagine a scenario where a railway company needs a modern  and efficient system  to handle their ticketing and management operiations.The current manual process is time - consuming ,prone to errors ,and lacks flexibility.They want to automate these processes and provide a convenent platform for their customers to book tickets,check fares ,and manage their reservaations easily.

The Data Analysi part of the RAILWAY MANAGEMENT SYSTEM involves analysing the data collected from various sources.such as customer bookings,train schedules, and fare information.This analysis helps in gaining  insights into customer prefferences,popular routes etc.

>Objective:

The primary objective of the Railway management system is to  streamline and automate the ticketing and management process for both the railway company and its  customers.The specific objectives include :
 

Ticket booking: ALLOW users to book tickets convinently by selecting th desired train,class,and travel date .The system should validate the availability  of seats and provide a confirmation for successful bookings

>Reservation cancellation:

Provide users with the option  to cancel their booked tickets if requires.The should handle cancellations requests,update seat availability,and process refunds if applicable.

>Fare checking:

Enable users to check fares for different trains, classes, and routes before makinng a booking,The system should fetch  the fare deatails fom the database and display them accurately.

>Booking History:

Allow users to view their booking history.including details such as train name,depature?arrival times,class and booking status.

This feature helps users keep track of their travel history.


>Train Availability:Provide users with an up to date list of availabaility  trains based on the selected source and destination.The system should display relevennt details like train name,departure/arrival times,and seat availability.

Menu: Display a menu to users,showcasing  the available options and functionalities of the system .
>Exit:Allow users to  exit the program gracefully when they have completed their tasks.




```python
import datetime

# Sample data for trains and fares
trains = [
    {"name": "Express 101", "departure": "08:00", "arrival": "12:00", "seats_available": 100},
    {"name": "Local 201", "departure": "10:00", "arrival": "14:00", "seats_available": 50},
    {"name": "Superfast 301", "departure": "12:00", "arrival": "16:00", "seats_available": 75},
]

fares = {"Economy": 50, "Business": 100}

# Sample data for booked tickets
booked_tickets = []

def display_menu():
    print("Railway Management System Menu:")
    print("1. Book Ticket")
    print("2. Cancel Booking")
    print("3. Check Fare")
    print("4. Booking History")
    print("5. Train Availability")
    print("6. Exit")

def book_ticket():
    print("\nBook Ticket:")
    print("Available Trains:")
    for i, train in enumerate(trains, 1):
        print(f"{i}. {train['name']} ({train['departure']} - {train['arrival']})")

    try:
        choice = int(input("Enter the number of the train you want to book: ")) - 1
        if 0 <= choice < len(trains):
            train = trains[choice]
            print(f"Selected train: {train['name']}")
            class_choice = input("Enter class (Economy/Business): ").capitalize()
            if class_choice in fares:
                if train['seats_available'] > 0:
                    ticket = {
                        "train_name": train['name'],
                        "class": class_choice,
                        "booking_time": datetime.datetime.now(),
                    }
                    booked_tickets.append(ticket)
                    train['seats_available'] -= 1
                    print("Booking confirmed!")
                else:
                    print("Sorry, no seats available for this train.")
            else:
                print("Invalid class choice.")
        else:
            print("Invalid train choice.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")

def cancel_booking():
    print("\nCancel Booking:")
    if len(booked_tickets) > 0:
        for i, ticket in enumerate(booked_tickets, 1):
            print(f"{i}. {ticket['train_name']} ({ticket['class']})")
        
        try:
            choice = int(input("Enter the number of the ticket you want to cancel: ")) - 1
            if 0 <= choice < len(booked_tickets):
                canceled_ticket = booked_tickets.pop(choice)
                for train in trains:
                    if train['name'] == canceled_ticket['train_name']:
                        train['seats_available'] += 1
                        print("Booking canceled successfully.")
                        break
            else:
                print("Invalid ticket choice.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    else:
        print("No booked tickets to cancel.")

def check_fare():
    print("\nCheck Fare:")
    class_choice = input("Enter class (Economy/Business): ").capitalize()
    if class_choice in fares:
        print(f"Fare for {class_choice} class: ${fares[class_choice]}")
    else:
        print("Invalid class choice.")

def booking_history():
    print("\nBooking History:")
    if len(booked_tickets) > 0:
        for i, ticket in enumerate(booked_tickets, 1):
            print(f"{i}. Train: {ticket['train_name']}, Class: {ticket['class']}, Booking Time: {ticket['booking_time']}")
    else:
        print("No booking history.")

def train_availability():
    print("\nTrain Availability:")
    for train in trains:
        print(f"Train: {train['name']}, Departure: {train['departure']}, Arrival: {train['arrival']}, Seats Available: {train['seats_available']}")

while True:
    display_menu()
    choice = input("Enter your choice: ")
    
    if choice == '1':
        book_ticket()
    elif choice == '2':
        cancel_booking()
    elif choice == '3':
        check_fare()
    elif choice == '4':
        booking_history()
    elif choice == '5':
        train_availability()
    elif choice == '6':
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid choice. Please select a valid option.")
```

```python
from datetime import datetime, timedelta

class Car:
    def __init__(self, car_id, manufacturer, model, car_type, age, km_driven, last_service_date, service_history):
        self.car_id = car_id
        self.manufacturer = manufacturer
        self.model = model
        self.car_type = car_type  # Basic, Mid-range, Luxury
        self.age = age
        self.km_driven = km_driven
        self.last_service_date = last_service_date
        self.service_history = service_history

    def __str__(self):
        return f"{self.manufacturer} {self.model} ({self.car_type}) - Age: {self.age} years, " \
               f"Kilometers Driven: {self.km_driven}, Last Service: {self.last_service_date}"

class CarRentalSystem:
    def __init__(self):
        self.cars = []
        self.rented_cars = {}

    def add_car(self, car):
        self.cars.append(car)

    def remove_car(self, car_id):
        for car in self.cars:
            if car.car_id == car_id:
                self.cars.remove(car)
                return True
        return False

    def rent_car(self, car_id, days, extra_kms=0):
        for car in self.cars:
            if car.car_id == car_id:
                rental_cost = self.calculate_rental_cost(car, days, extra_kms)
                self.rented_cars[car_id] = {"car": car, "days": days, "extra_kms": extra_kms, "cost": rental_cost}
                return rental_cost
        return None

    def return_car(self, car_id):
        if car_id in self.rented_cars:
            return self.rented_cars.pop(car_id)
        return None

    def calculate_rental_cost(self, car, days, extra_kms):
        base_rate = self.get_base_rate(car)
        extra_km_rate = 5  # Adjust as needed
        total_cost = base_rate * days + extra_km_rate * extra_kms
        return total_cost

    def get_base_rate(self, car):
        car_type_rates = {"Basic": 30, "Mid-range": 60, "Luxury": 100}
        return car_type_rates.get(car.car_type, 0)

def main():
    rental_system = CarRentalSystem()

    car1 = Car(1, "Toyota", "Corolla", "Basic", 2, 20000, "2023-01-01", "Service history available")
    car2 = Car(2, "Mercedes-Benz", "S-Class", "Luxury", 1, 15000, "2023-02-15", "Service history available")

    rental_system.add_car(car1)
    rental_system.add_car(car2)

    while True:
        print("\nCar Rental Menu:")
        print("1. Rent a Car")
        print("2. Return a Car")
        print("3. List Available Cars")
        print("4. Exit")
        
        choice = input("Enter your choice: ")

        if choice == '1':
            car_id = int(input("Enter the car ID you want to rent: "))
            days = int(input("Enter the number of days for rental: "))
            extra_kms = int(input("Enter extra kilometers driven (if any): "))
            cost = rental_system.rent_car(car_id, days, extra_kms)
            if cost is not None:
                print(f"Rental cost: ${cost}")
            else:
                print("Car not found.")
        elif choice == '2':
            car_id = int(input("Enter the car ID you want to return: "))
            rental_info = rental_system.return_car(car_id)
            if rental_info:
                print(f"Returned car: {rental_info['car']}")
                print(f"Rental cost: ${rental_info['cost']}")
            else:
                print("Car not found.")
        elif choice == '3':
            print("\nAvailable Cars:")
            for car in rental_system.cars:
                print(car)
        elif choice == '4':
            print("Exiting the Car Rental System. Thank you!")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('youtube_data.csv')  # Replace 'youtube_data.csv' with the actual dataset file name

# Display basic information about the dataset
print("Dataset Information:")
print(df.info())

# Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Explore the distribution of views
plt.figure(figsize=(10, 6))
sns.histplot(df['views'], bins=50, kde=True)
plt.title("Distribution of Views")
plt.xlabel("Views")
plt.ylabel("Frequency")
plt.show()

# Explore the correlation between variables
plt.figure(figsize=(10, 6))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title("Correlation Heatmap")
plt.show()

# Explore the relationship between likes and dislikes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='likes', y='dislikes', hue='category_id')
plt.title("Likes vs Dislikes")
plt.xlabel("Likes")
plt.ylabel("Dislikes")
plt.show()

# Explore the distribution of comment counts
plt.figure(figsize=(10, 6))
sns.boxplot(x='comment_disabled', y='comment_count', data=df)
plt.title("Comment Disabled vs Comment Count")
plt.xlabel("Comment Disabled")
plt.ylabel("Comment Count")
plt.show()

# Explore the relationship between tags and views
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Tag appeared in title', y='views', hue='like_dislike_disabled')
plt.title("Tag Appeared in Title vs Views")
plt.xlabel("Tag Appeared in Title")
plt.ylabel("Views")
plt.show()
```
