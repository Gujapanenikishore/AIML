# 1.write a python program to find a target values in a list using linear search with following steps:
a.Initialize the list to store the input elements.
2.Initialize found=false
c.ente the item to be searched (match_item)
d.for each element in the list
   1.if match_item=value
   a return match_item's position.
e. if the match_item is not in the list,display an error message that the item is not found in the list.




```python
def linear_search(input_list, match_item):
    found = False
    position = None

    for index, value in enumerate(input_list):
        if value == match_item:
            found = True
            position = index
            break

    return position

# Initialize the list with input elements
input_list = [12, 45, 67, 23, 56, 89, 34, 78]

# Input the item to be searched
match_item = int(input("Enter the item to be searched: "))

# Perform linear search
result = linear_search(input_list, match_item)

if result is not None:
    print(f"{match_item} found at position {result}")
else:
    print(f"{match_item} is not found in the list.")
```

# 2.write a python program to implement binary search  to find the target values  from the list:
a.create a separate function to do binary search.
b.get the no of inputs from the user
c.store the inputs individually in a list
d.in binary search function at fist sort the list in order to  start teh search from middle of the list
e.compare the middle element  to right and left elements to search target element
f.if greaterr,move to right of list or else move on aanother side of the list.
g.print the result along with the position of the element.

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

# Get the number of inputs from the user
num_inputs = int(input("Enter the number of inputs: "))

# Initialize an empty list to store the inputs
input_list = []

# Input the elements individually and add them to the list
for i in range(num_inputs):
    element = int(input(f"Enter element {i + 1}: "))
    input_list.append(element)

# Sort the list in ascending order for binary search
input_list.sort()

# Input the target element to be searched
target_element = int(input("Enter the target element to be searched: "))

# Perform binary search
result = binary_search(input_list, target_element)

if result != -1:
    print(f"{target_element} found at position {result}")
else:
    print(f"{target_element} is not found in the list.")
```

# 3.write a python program for sorting a list of elements  using selection sort algorithm:
a,assume two list: sorted list-intially empty and unsorted list given input list
b.in the first  iteration,find the smallest  element in the unsorted list and place it in the sorted list
c.in the second iteration,find the smallest  element in the unsorted list and place it in the correct position by comparing with the element in the sorted list.
d.in the third iteration,again find the smallest  element in the unsorted list and place it in the correct position by comparing with the  elements in the sorted list
e.this  process continues till the unsorted list becomes empty.
f.display the sorted list



```python
def selection_sort(input_list):
    sorted_list = []

    while input_list:
        min_value = min(input_list)  # Find the smallest element in the unsorted list
        input_list.remove(min_value)  # Remove the smallest element from the unsorted list
        sorted_list.append(min_value)  # Append it to the sorted list

    return sorted_list

# Input the list of elements
input_list = [64, 25, 12, 22, 11]

# Call the selection_sort function to sort the list
sorted_list = selection_sort(input_list)

# Display the sorted list
print("Sorted list:", sorted_list)
```

# 4.write a python program for sorting a list of elements using insertion sort alogorithm.
a,assume two list lists :  sorted list - intially empty and unsorted list given input list.
b.in the first iteration ,take the first  element in the unsorted list and insert it in  sorted list.
c.in the second iteration,take the second element in the given list and compare with the element in the sorted sublist and place it in the correct position.
d.in the third iteration ,take the third element in the gib=ven list and compare with the elements in the sorted  sublist and place  the element in the correct position.
e.these process continues untill the last element is inserted in the sorted sublist
f.display the sorted elements


```python
def insertion_sort(input_list):
    for i in range(1, len(input_list)):
        key = input_list[i]  # Get the current element to be inserted
        j = i - 1

        # Compare the current element with elements in the sorted list and shift them to the right if necessary
        while j >= 0 and key < input_list[j]:
            input_list[j + 1] = input_list[j]
            j -= 1

        input_list[j + 1] = key  # Insert the current element in the correct position

# Input the list of elements
input_list = [64, 25, 12, 22, 11]

# Call the insertion_sort function to sort the list
insertion_sort(input_list)

# Display the sorted list
print("Sorted list:", input_list)
```

# 5.wrte a python program that performs merge sort on the list of numbers:
a.Divide: if the given array has zero or one element,return
i.otherwise
iidivide the input list in to two halves each containing half of the elements i.e. left half and right half.
b.Conquer: recursively sort the two  two lists(left half and right half).
a.call the mergw sort on the left half.
b.call the merge sort on the right half.
c.combine: combine the elements  back in the input list by merging the two sorted  lists into a sorted sequences.


```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2  # Calculate the middle index

        # Divide the input list into two halves
        left_half = arr[:mid]
        right_half = arr[mid:]

        # Recursively sort the left and right halves
        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        # Merge the sorted halves back into the input list
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        # Check if any elements were left in the left and right halves
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

# Input the list of numbers
input_list = [38, 27, 43, 3, 9, 82, 10]

# Call the merge_sort function to sort the list
merge_sort(input_list)

# Display the sorted list
print("Sorted list:", input_list)
```

# 6.Write  a python script  to perform  the following operations on a singlly linked list
a.create a list
b.find the smallest element from the list 
c.insert an element  if it is not  a duplicate  element 
d.display  the elements in reverse order


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    # Function to insert a new element at the end of the linked list
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    # Function to find the smallest element in the linked list
    def find_smallest(self):
        if not self.head:
            return None
        current = self.head
        smallest = current.data
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest

    # Function to insert an element if it's not a duplicate
    def insert_if_not_duplicate(self, data):
        if not self.head:
            self.head = Node(data)
            return
        current = self.head
        while current:
            if current.data == data:
                return
            if current.next is None:
                current.next = Node(data)
                return
            current = current.next

    # Function to display the elements in reverse order
    def display_reverse(self):
        stack = []
        current = self.head
        while current:
            stack.append(current.data)
            current = current.next
        while stack:
            print(stack.pop(), end=" -> ")
        print("None")

# Create a singly linked list
my_list = LinkedList()
my_list.append(10)
my_list.append(5)
my_list.append(15)
my_list.append(5)

# Find the smallest element
smallest = my_list.find_smallest()
print(f"The smallest element is: {smallest}")

# Insert an element if it's not a duplicate
my_list.insert_if_not_duplicate(8)

# Display the elements in reverse order
print("Elements in reverse order:")
my_list.display_reverse()
```

# 7.write a python program to implement the various operaitons for stack ADT
i)push  ii).pop  iii).display


```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None  # Return None when trying to pop from an empty stack

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        if not self.is_empty():
            for item in reversed(self.items):
                print(item)
        else:
            print("Stack is empty.")

# Create a stack
stack = Stack()

# Push elements onto the stack
stack.push(10)
stack.push(20)
stack.push(30)

# Display the elements in the stack
print("Stack elements:")
stack.display()

# Pop an element from the stack
popped_item = stack.pop()
if popped_item is not None:
    print(f"Popped item: {popped_item}")

# Display the updated stack
print("Updated stack elements:")
stack.display()
```

# 8.write a python script to implement the various operiations for queue ADT
 i)insert ii).delete iii)display
 

```python
class Queue:
    def __init__(self):
        self.items = []

    def insert(self, item):
        self.items.append(item)

    def delete(self):
        if not self.is_empty():
            return self.items.pop(0)
        else:
            return None  # Return None when trying to dequeue from an empty queue

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        if not self.is_empty():
            for item in self.items:
                print(item)
        else:
            print("Queue is empty.")

# Create a queue
queue = Queue()

# Insert elements into the queue
queue.insert(10)
queue.insert(20)
queue.insert(30)

# Display the elements in the queue
print("Queue elements:")
queue.display()

# Delete an element from the queue
deleted_item = queue.delete()
if deleted_item is not None:
    print(f"Deleted item: {deleted_item}")

# Display the updated queue
print("Updated queue elements:")
queue.display()
```
